```python
def test_vin(vin):
    url_base = 'https://vpic.nhtsa.dot.gov/api/vehicles/DecodeVINValuesBatch/'
    form = '?format=json'
    url = url_base+vin+form
    return requests.get(url).json()
x = test_vin('3GTP1VEC4EG551563')
```

```python
x['Results']
```

```python
#return df from batch of 50 or less VIN #'s in VIN;VIN;VIN format
def batch_VIN(vinlis):
    url = 'https://vpic.nhtsa.dot.gov/api/vehicles/DecodeVINValuesBatch/'
    post_fields = {'format': 'json', 'data':vinlis}
    r = requests.post(url, data=post_fields)
    return pd.DataFrame.from_dict(json.loads(r.text)['Results'])
```

```python
def create_vin_strings(vin_lis):
    if len(vin_lis) < 51:
        stri = ';'.join(vin_lis)
        return stri
    else:
        return 'vin list too long'
        
```

```python
batch2 = all_vins[50: 100]
create_vin_strings(batch2)
```

```python
# 101928 is total number of vins
def create_master_list(lis):
    master = []
    counter = 0
    x = np.arange(0, len(lis), 50)
    for n in x:
        master.append(create_vin_strings(lis[counter: counter+50]))
        counter+= 50
    return master
masterlis = create_master_list(all_vins)
```

```python
def giant_df(big_lis):
    dflis = []
    for big in big_lis:
        dflis.append(batch_VIN(big))
    return pd.concat(dflis)
```

```python
gdf = giant_df(masterlis)
```
