```python
def scatter(xcol,ycol,df):
    
        # Scatter plot
    plt.figure(figsize=(8, 6))  # Optional: Adjust the size of the plot

    plt.scatter(df[xcol], df[ycol], color='blue', marker='o', label='Data')

    # Add labels and title
    plt.xlabel(xcol)
    plt.ylabel(ycol)
    plt.title('Scatter Plot of ' + ycol + ' vs ' + xcol)

    # Add grid if desired
    plt.grid(True)

    # Add legend
    plt.legend()

    # Show plot
    plt.show()
```

EngineHP has stronger correlation with eventual list price than any other feature. Surprising! It is likely the fact that EngineHP has ~50% missing values will affect the true strength of this feature. Year and Odometer have NO missing values and show a correlation of **0.55** (year/price corr) and and **-0.53** (odometer/price), which are strong. These features themselves are highly correlated with a score of **-0.63**, as expected. However, **DisplacementCC and EngineHP have the strongest correlation with 0.769**, even more than year and oodometer's score of -0.63, yet **DisplacementCC has 0.4 correlation with price while EngineHP has 0.612 correlation with price. Therefore, DisplacementCC will likely *not* be chosen in the eventual model.** TopSpeedMPH is not very useful, showing very weak correlation with price, and slight correlation with Year. It is likely that the feature of year and odometer will need to be counted as one feature, while **BasePrice** is able to be worked into the eventual model, showing a strong correlation with price of **.628**. It is a good idea to try to get an idea of which cars are luxury cars made with premium materials or big trucks which have a much higher MSRP. I am still working on inputting the base MSRP to the data, and other features such as EngineHP and the categorical data, could be scraped as well as this base MSRP. For the features above, once some tests were run to determine their viability, I would predict features chosen to be odometer, year, and EngineHP. It will be interesting to see how different features affect the depreciation of each vehicle and which feature strengths may change depending on other features of the vehicle. For example, convertibles likely do not hold their value as well as a truck, which provides utility in moving large objects.
