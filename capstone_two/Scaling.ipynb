```python
useful_categoricals = ['GVWR', 'ABS', 'AirBagLocCurtain', 'AirBagLocSide', 'BodyClass', 'BrakeSystemType','Doors', 'DriveType', 
        'DynamicBrakeSupport', 'ESC', 'EngineConfiguration', 'EngineCylinders', 'FuelInjectionType','FuelTypePrimary',
        'KeylessIgnition', 'RearVisibilitySystem', 'ValveTrainDesign', 'Turbo', 'ValveTrainDesign']

less_useful_categoricals = ['AdaptiveCruiseControl', 'AdaptiveDrivingBeam', 'AirBagLocKnee', 'AutoReverseSystem','BlindSpotMon',
            'CAN_AACN', 'CIB', 'EngineCycles', 'ForwardCollisionWarning', 'FuelTypeSecondary',
            'LaneDepartureWarning', 'LaneKeepSystem', 'LowerBeamHeadlampLightSource',  'PedestrianAutomaticEmergencyBraking',
            'Pretensioner', 'SeatRows', 'Seats', 'SemiautomaticHeadlampBeamSwitching', 'TractionControl', 'TransmissionSpeeds',
            'TransmissionStyle', 'WheelSizeFront', 'WheelSizeRear', 'Windows']

veh_indicators = ['Make', 'Model', 'Series','Series2','Trim','Trim2',
    'VehicleDescriptor', 'VehicleType', 'VIN']

number_cols = ['BasePrice', 'DisplacementCC', 'EngineHP', 'Year', 'odometer']

target=['price']
```

```python
dummies = useful_categoricals + less_useful_categoricals
df_cat = df[dummies]
df_no_cat = df[number_cols]
```

```python
print(df_cat.columns.to_list() == dummies)
print(df_no_cat.columns.to_list() == number_cols)
```

```python
df_dummies = pd.get_dummies(df_cat.astype(str))
df_dummies.columns.to_list()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() 
scaled_df = scaler.fit_transform(df_no_cat) 
scaled_df = pd.DataFrame(scaled_df, columns=number_cols) 

scaled_df.head()
```

```python
new_df = pd.concat([scaled_df, df_dummies, df['price']], axis=1)
new_df.shape
```

```python
from sklearn.model_selection import train_test_split

X = new_df.drop(['price'], axis=1)
y = new_df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```
